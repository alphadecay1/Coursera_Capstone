```python
import pandas as pd
import numpy as np

%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
```

```python
#Read the data in a data frame
df = pd.read_csv('data.csv')
df.head()
```

Removing all the not required attributes from the data frame.

```python
df.drop(['Accident_Index', 'Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude', 'Latitude', 'Date', 'Time', 'Local_Authority_.Highway.', 'Local_Authority_.District.', 'X1st_Road_Number', 'X2nd_Road_Number', 'Did_Police_Officer_Attend_Scene_of_Accident', 'LSOA_of_Accident_Location'], axis=1, inplace=True)
df.head()
```

```python
attributes=["Number_of_Vehicles","Number_of_Casualties", "Road_Type", "Pedestrian_Crossing.Human_Control", "Pedestrian_Crossing.Physical_Facilities", "Light_Conditions", "Weather_Conditions", "Road_Surface_Conditions","Accident_Severity"]
df = df[attributes]

#splitting Attributes into two groups for better plotting

attributes1=["Number_of_Vehicles","Number_of_Casualties", "Road_Type", "Pedestrian_Crossing.Human_Control","Accident_Severity"]
attributes2=["Pedestrian_Crossing.Physical_Facilities", "Light_Conditions", "Weather_Conditions", "Road_Surface_Conditions","Accident_Severity"]
```

```python
#scatter matrix for second group of attributes
scatter_matrix(df[attributes2], figsize=(15, 15))
```

```python
# plotting the correlation matrix 

plt.matshow(df.corr())
plt.show()
```

```python
corr_matrix = df.corr()
corr_matrix["Accident_Severity"].sort_values(ascending=False)
```

```python
# defining a feature set for the classifier algorithms

feature_set=["Number_of_Vehicles","Number_of_Casualties", "Road_Type", "Pedestrian_Crossing.Human_Control", "Pedestrian_Crossing.Physical_Facilities", "Light_Conditions", "Weather_Conditions", "Road_Surface_Conditions"]
```

```python
# randomly selecting samples for the training and testing the models
subdf = df.sample(n=100000)
```

```python
X = np.asarray(subdf[feature_set])
```

```python
y=np.asarray(subdf['Accident_Severity'])
```

```python
X[0:5]
```

```python
y[0:5]
```

```python
# Splitting the data into training and testing set

X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.25, random_state=4)
```

K-Map Classifier

```python
#finding the best K
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
Ks = 10
mean_acc = np.zeros((Ks-1))
std_acc = np.zeros((Ks-1))
ConfustionMx = [];
for n in range(1,Ks):
    
    #Train Model and Predict  
    neigh = KNeighborsClassifier(n_neighbors = n).fit(X_train,y_train)
    yhat=neigh.predict(X_test)
    mean_acc[n-1] = metrics.accuracy_score(y_test, yhat)

    
    std_acc[n-1]=np.std(yhat==y_test)/np.sqrt(yhat.shape[0])

mean_acc
plt.plot(range(1,Ks),mean_acc,'g')
plt.fill_between(range(1,Ks),mean_acc - 1 * std_acc,mean_acc + 1 * std_acc, alpha=0.10)
plt.legend(('Accuracy ', '+/- 3xstd'))
plt.ylabel('Accuracy ')
plt.xlabel('Number of Nabors (K)')
plt.tight_layout()
plt.show()
```

```python
k = 7
neigh7 = KNeighborsClassifier(n_neighbors = k).fit(X_train,y_train)
yhat7 = neigh7.predict(X_test)
print("Train set Accuracy: ", metrics.accuracy_score(y_train, neigh7.predict(X_train)))
print("Test set Accuracy: ", metrics.accuracy_score(y_test, yhat7))
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth = 4)
drugTree.fit(X_train,y_train)
predTree = drugTree.predict(X_test)
print("DecisionTrees's Accuracy: ", metrics.accuracy_score(y_test, predTree))
```

SVM

```python
from sklearn import svm
clf = svm.SVC(kernel='rbf', probability=True)
clf.fit(X_train, y_train) 
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
yhatLR = LR.predict(X_test)
```

Model Evaluation

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import log_loss

yhatKmap = neigh7.predict(X_test)
yhatKmap_prob = neigh7.predict_proba(X_test)
yhatDtree = drugTree.predict(X_test)
yhatDtree_prob = drugTree.predict_proba(X_test)
yhatSVM = clf.predict(X_test)
yhatSVM_prob = clf.predict_proba(X_test)
yhatLR = LR.predict(X_test)
yhatLR_prob = LR.predict_proba(X_test)

f1_score_Kmap = f1_score(y_test, yhatKmap, average='weighted') 
jaccard_Kmap = jaccard_score(y_test, yhatKmap, average='weighted')
LogLoss_Kmap = log_loss(y_test, yhatKmap_prob)
f1_score_Dtree = f1_score(y_test, yhatDtree, average='weighted') 
jaccard_Dtree = jaccard_score(y_test, yhatDtree, average='weighted')
LogLoss_Dtree = log_loss(y_test, yhatDtree_prob)
f1_score_SVM = f1_score(y_test, yhatSVM, average='weighted') 
jaccard_SVM = jaccard_score(y_test, yhatSVM, average='weighted')
LogLoss_SVM = log_loss(y_test, yhatSVM_prob)
f1_score_LR = f1_score(y_test, yhatLR, average='weighted') 
jaccard_LR = jaccard_score(y_test, yhatLR, average='weighted')
LogLoss_LR = log_loss(y_test, yhatLR_prob)
```

```python
data = {'Algorithm':['KNN', 'Decision Tree', 'SVM', 'Logistic Regression'],
        'Jaccard':[jaccard_Kmap, jaccard_Dtree, jaccard_SVM, jaccard_LR],
        'F1-score':[f1_score_Kmap, f1_score_Dtree, f1_score_SVM, f1_score_LR],
        'LogLoss':[LogLoss_Kmap, LogLoss_Dtree, LogLoss_SVM, LogLoss_LR]
        }

dfd = pd.DataFrame (data, columns = ['Algorithm','Jaccard', 'F1-score', 'LogLoss'])
dfd
```
